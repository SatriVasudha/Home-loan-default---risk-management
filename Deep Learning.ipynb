### Importing libraries

```python
import pandas as pd
import numpy as np

# featuretools for automated feature engineering
#import featuretools as ft

# matplotlib and seaborn for visualizations
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 22
import seaborn as sns

# Suppress warnings from pandas
import warnings
warnings.filterwarnings('ignore')

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest 
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers.normalization import BatchNormalization
```

### Fetching final merged file

```python
app_train = pd.read_csv('train_final.csv').sort_values('SK_ID_CURR').reset_index(drop = True)
```

```python
app_train
```

```python
# Independent variable
x=app_train.drop(columns = ['TARGET','Unnamed: 0'])
```

```python
x
```

```python
# Dependent variable
y=app_train.iloc[:,2].values
```

```python
y
```

```python
# Function to check missing values
def missing_values_table(df):
    mis_val=df.isnull().sum()
    mis_val_percent=100*mis_val/len(df)
    mis_val_table=pd.concat([mis_val,mis_val_percent],axis=1)
    mis_val_table_ren_columns=mis_val_table.rename(columns={0:'Missing values',1:'% of Total_values'})
    mis_val_table_ren_columns=mis_val_table_ren_columns[mis_val_table_ren_columns.iloc[:,1]!=0].sort_values('% of Total_values',ascending=False).round(1)
    return mis_val_table_ren_columns
```

```python
missing_x=missing_values_table(x)
missing_x
```

```python
# Imputing missing values with mean
imputer=SimpleImputer(strategy='mean')
imputer.fit(x)
x=imputer.transform(x)
```

```python
# Converting inf values to zero
x[x<0]=0
```

```python
# Feature selection
chi2_features = SelectKBest(chi2, k = 1000) 
X_kbest_features = chi2_features.fit_transform(x, y) 
```

```python
print('Original feature number:', x.shape[1]) 
print('Reduced feature number:', X_kbest_features.shape[1])
```

```python
# Splitting data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

### Deep Learning

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
x_test=np.array(x_test)
y_test=np.array(y_test)
```

### Deep Neural network with 4 layers

```python
#4 layers with 2 hidden layer and the hidden layer has 80 neurons 
```

```python
model =keras.Sequential([
    keras.layers.Flatten(input_shape=(1185,)),
    keras.layers.Dense(80, activation=tf.nn.relu),
	keras.layers.Dense(80, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid),
])
```

```python
model.compile(optimizer='adam', 
              loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=50, batch_size=1)
```

```python
model.summary()
```

```python
test_loss, test_acc = model.evaluate(x_test,y_test)

print('Test accuracy:', test_acc)
```

```python
predictions = model.predict(x_test)
```

```python
accuracy_score(y_test, predictions.round()) 
```

### Deep Neural network with 5 layers

```python
# 5 layers with 3 hidden layers, first and second hidden layer has 80 neurons and third hidden layer has 40 neurons
```

```python
model =keras.Sequential([
    keras.layers.Flatten(input_shape=(1185,)),
    keras.layers.Dense(80, activation=tf.nn.relu),
    keras.layers.Dense(80, activation=tf.nn.relu),
    keras.layers.Dense(40, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid),
])
```

```python
model.compile(optimizer='adam', 
              loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=50, batch_size=1)
```

```python
model.summary()
```

```python
test_loss, test_acc = model.evaluate(x_test,y_test)

print('Test accuracy:', test_acc)
```

```python
predictions = model.predict(x_test)
```

```python
accuracy_score(y_test, predictions.round()) 
```
